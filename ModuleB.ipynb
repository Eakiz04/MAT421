17.1 Interpolation Problem Statement
*   **Interpolation** is the process used to estimate unknown values using known data values
*   For a data set consisting of independent data values, $x_i$, and dependent data values, $y_i$, we can find an estimation function $\hat{y}(x)$ such that $\hat{y}(x_i)=y_i$. Where $\hat{y}(x)$ is the **interpolation function**
 for every point in our data set
*   Interpolation vs Regression: Interpolation does not require finding an underlying model for the data like regression does. Interpolation works better for a data set that contains an accurate data range



```python
import numpy as np
import matplotlib.pyplot as plt
x = [1, 3, 5, 7]
y = [1.5, 3, 2.5, 3.2]

print("Here we want to know what y is when x = 4")

# Interpolation problem statement figure
plt.figure(figsize = (5,5))
plt.plot(x, y, 'bo')
plt.text(4, 2.5, "?", fontsize=20, color="red", ha='center', va='center')
plt.title('Interpolation Propblem Statement')
plt.ylim(0,5)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

17.2 Linear Interpolation
*   The estimated point is assumed to lie on the line joining the nearest points to the left and right
*   linear interpolation at x: $\hat{y}(x) = y_i + \frac{(y_{i+1}-y_i)(x-x_i)}{(x_{i+1}-x_i)}$
*   Example: For linear interpolation at x = 4 based on data x = [1,3,5,7], y = [1.5,3,2.5,3.2]

  3 < x < 5 thus,

  $\hat{y}(x) = y_i + \frac{(y_{i+1}-y_i)(x-x_i)}{(x_{i+1}-x_i)}=3+\frac{(2.5-3)(4-3)}{5-3}=2.75$


```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
x = [1, 3, 5, 7]
y = [1.5, 3, 2.5, 3.2]

f = interp1d(x, y)
# Print y when x = 4
print("y =", end=" ")
print(f(4), end=" ")
print("when x = 4")

# Interpolation problem statement figure
plt.figure(figsize = (5,5))
plt.plot(x, y, '-bo')
plt.plot(4, f(4), 'ro')
plt.title('Linear Interpolation at x = 4')
plt.ylim(0,5)
plt.xlabel('x')
plt.ylabel('y')
plt.show()

2.75 == f(4)
```

17.3
*    a set of piecewise cubic functions. Assuming that the points $(x_i,y_i)$ and $(x_{i+1},y_{i+1})$ are joined by cubic polynomial $S_i(x)=a_ix^3+b_ix^2+c_ix+d_i$ that is valid for $x_i \leq x \leq x_{i+1}$ for i = 1, ... , n-1



```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt

x = [1, 3, 5, 7]
y = [1.5, 3, 2.5, 3.2]

# Find y when for 1 <= x <= 7
f = CubicSpline(x, y, bc_type='natural')
x_new = np.linspace(1, 7, 100)
y_new = f(x_new)

# Interpolation problem statement figure
plt.figure(figsize = (10,8))
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
