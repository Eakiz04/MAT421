3.2 Continunity and Differentiation

*   **Limits:** Used to define continuity, derivatives, and integral
*   **Euclidean norm:** $||x|| = \sqrt{\sum_{i=1}^dx_i^2}$ for $x=(x_1,...,x_d)^T \in ℝ^d$
*   **Limits of a function:** Let $f:D → ℝ$ be a real-valued function on $D⊆ℝ^d$. Then $f$ is said to have limit $L\in ℝ$ as x approaches $a$ if for any $ϵ>0$, there exists a $δ>0$ such that $|f(x)-L| < ϵ$ for all $x \in D \cap B_\delta (a) /${a}. This is written as:  

$\hspace{8.5cm}\lim_{x \rightarrow a} f(x)=L$
*   **Continunity:** A Continuous function is one that doesn't have any abrupt changes in value or discontinuities. $f$ is continuous at $a \in D$ if

$\hspace{8.5cm}\lim_{x \rightarrow a} f(x)=f(a)$
*   $h=g \circ f ⇒ h(x) = g(f(x))$
*   Assuming that $f$ is continuous at $x_0$ and that $g$ is continous $f(x_0)$. Then $ g \circ f$ is continous at $x_0$

*   **Derivative:** Measures the sensitivity to change of the function value (output value) with respect to another variable. Let $f:D \rightarrow ℝ$ where $D \subseteq ℝ$ and let $x_0
\in D$ be an interior point of D. The derivative of $f$ at $x_0$ is

$\hspace{8.5cm}f'(x_0)= \frac{df(x_0)}{dx} = \lim_{h \rightarrow 0} \frac{f(x_0+h) - f(x_0)}{h}$

Example 1: Continuous vs non-continous

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-5,5,200)
y1= np.sin(1/x) # Non-Continous
y2= np.cos(2*x) # Continuous

plt.figure
plt.plot(x,y1,'-r')
plt.plot(x,y2,'-b')
plt.title("Continuous (Blue) vs Non-Continuous (Red)")
```

Example 2: Derivative at $x_0=2$ when $f(x) = 3x^2 + 5x + 7$

```python
import sympy as sp

# Example: Derivative of f(x0)
x = sp.symbols('x')
f = 3*x**2 + 5*x + 7
x0 = 2

# Derivative using sympy diffferential function
df = sp.diff(f,x)
ans1 = df.subs(x,x0)

# Derivative using limit formula
h = 1e-6
ans2 = (f.subs(x, 2 + h) - f.subs(x, x0)) / h

print("Derivative f(x0)' where x0=2:")
print(f"f(x0)' = {ans1}")
print(f"f(x0)' = {ans2}")
```

Taylor Theorem
*   Gives an approximation of a differentiable function around
a given point by a polynomial
*   Let $f:D \rightarrow ℝ$ where $D ⊆ ℝ$. Suppose $f$ has a $m$ times continuous derivative on $[a,b]$. Then
$f(b)=f(a)+f(b-a)f'(a)+\frac{1}{2}(b-a)^2f''(a)+...+\frac{(b-a)^{m-1}}{(m-1)!}f^{(m-1)}(a)+R_m$, where $R_m=\frac{(b-a)^m}{(m)!}f^m(a+ \theta(b-a))$ for some $0 < \theta < 1$.




Example 3: Taylor Theorem

```python
import sympy as sp
import matplotlib.pyplot as plt

x = sp.symbols('x')
f = sp.sin(x)
x0 = 0
n = 5
t = 0

# Compute the Taylor series expansion up to n terms at x0
i = 0
while i < n:
    t = t + sp.diff(f, x, i).subs(x, x0) * (x - x0)**i / sp.factorial(i)
    i = i + 1

# Print the Taylor series
print(f"Taylor series: {t}")

xp = np.linspace(-2, 2, 400)
plt.figure
plt.plot(xp, [t.subs(x, val) for val in xp],'--b')
plt.plot(xp, [f.subs(x, val) for val in xp],'-r')
plt.title("Taylor Series vs. Function")
```

3.3 Unconstrained Optimization
*   $\min_{x \in ℝ^d}f(x)$
*   **Global Minimizer:** Let $f:ℝ^d \rightarrow ℝ$. The point $x* \in ℝ^d$ is a global minimizer of $f$ over $ℝ^d$ if

$\hspace{8.5cm}f(x) \geq f(x*)$, $\forall x \in ℝ ^d$
*   **Local Minimizer:** Let $f:ℝ^d \rightarrow ℝ$. The point $x* \in ℝ^d$ is a local minimizer of $f$ over $ℝ^d$ if there is $\delta > 0$ such that

$\hspace{8.5cm}f(x) \geq f(x*)$, $\forall x \in B_\delta(x^*) /${x*}
*   **Decent Direction:** Let $f:ℝ^d \rightarrow ℝ$. A vector $v$ is a descent direction for $f$ at $x_0$ if there is $α^* > 0$ such that

$\hspace{8.5cm}f(x_0 + \alpha v) < f(x_0)$, $\forall \alpha \in (0,\alpha ^*)$
*   **Convexity:** A real-valued function is called convex if the line segment between any two points on the graph of the function lies above the graph between the two points
*   **Convex Set:** A set $D \subseteq ℝ^d$ is convex if $\forall x,y \in D$ and all $\alpha \in [0,1]$

$\hspace{8.5cm}(1- \alpha)x + \alpha y \in D$
* **Convex Function:** A function $f:ℝ^d \rightarrow ℝ$ is convex if, $\forall x,y \in ℝ^d$ and all $\alpha \in [0,1]$

$\hspace{8.5cm}f((1- \alpha)x + \alpha y) \leq (1-\alpha)f(x) + \alpha f(y)$
* **Gradient Descent:** An iterative optimization algorithm for finding a local
minimum of a differentiable function
*   **Steepest Descent:** Let $f:ℝ^d \rightarrow ℝ$ be continuously differential at $x_0$. For any unit vector $v \in ℝ^d$,

$\hspace{8.5cm}\frac{\partial f(x_0)}{\partial v} \geq \frac{\partial f(x_0)}{\partial v^*}$, where $v^* =- \frac{\nabla f(x_0)}{||\nabla f(x_0)||}$

Example 4: Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,5,200)
f  = x**4 - 5*x**2
x0 = 0
a = np.zeros(10)
a[0] = x0

# Gradient
for i in range(1, 10):
  g = 4 * x0**3 - 10 * x0
  x0 = x0 - .1*g
  a[i] = x0

plt.figure
plt.plot(x, f, '--b')
plt.plot(a,[(val)**4 - 5*(val)**2 for val in a], 'or')
```
